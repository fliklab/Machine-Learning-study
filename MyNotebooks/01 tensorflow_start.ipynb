```python
a = tf.Variable(3, name="input_a")
b = tf.Variable(5, name="input_b")
c = tf.multiply(a,b,name = "mul_c")
d = tf.add(a,b, name = "add_d")
e = tf.add(c,d, name ="add_e")
```

mul, sub, neg는 아래와 같이 변경되었다. 

tf.mul -> tf.multiply

tf.sub -> tf.subtract

tf.neg -> tf.negative

```python
sess = tf.Session() #tensorflow 세션 생성
init = tf.global_variables_initializer() #초기화 연산 정의(?)
sess.run(init) #초기화
```

init = tf.initialize_all_variables()
