## 평균 구하기 + plot으로 나타내기

```python
import matplotlib.pyplot as plt

#set을 입력하면, 평균을 리턴하는 함수 정의.
def mean(sampleset):
    total = 0
    for element in sampleset:
        total = total + element
    return total/len(sampleset)

myset = [-2., -10., 3., 6., 4., 6., 10.]
mymean = mean(myset)
plt.plot(myset)
plt.plot([mymean]*7)

plt.show()
```

```python
#위 코드에서 발전시켜봄.

%matplotlib inline 
 
import matplotlib.pyplot as plt
from numpy.random import randn
 
def mean(sampleset):
    total = 0
    for element in sampleset:
        total = total + element
    return total/len(sampleset)    

rand = randn(50)
rand_cumsum = rand.cumsum()
rand_mean = mean(rand)
rand_cumsum_mean= mean(rand_cumsum)

fig = plt.figure()

ax = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

ax.plot(rand) 
ax.plot([rand_mean]*len(rand))

ax2.plot(rand_cumsum,'k') 
ax2.plot([rand_cumsum_mean]*len(rand_cumsum),'r--') 

plt.show()


# fig : 는 figure이다. plt.figure
# ax : 서브플롯. fig에 ax라는 subplot을 추가(add_subplot)한다. subplot은 위치를 3개의 숫자로 지정해준다.
# ax.plot(----) : 서브플롯에서 그래프로 나타낼 데이터와 스타일을 입력변수로 넣어준다.
```

## 분산 구하기

```python
import matplotlib.pyplot as plt
import math
```

```python
#평균 구하기
def mean(sampleset):
    total = 0
    for element in sampleset:
        total = total + element
    return total/len(sampleset)

#분산 구하기
def variance(sampleset):
    total = 0
    setmean = mean(sampleset)
    for element in sampleset:
        #(값 - 평균)의 제곱의 평균
        total = total + (math.pow(element-setmean, 2))
    return total/len(sampleset)
```

```python
myset1 = [2., 10. , 3., 6., 4., 6., 10.]
myset2 = [1. , -39, 15, -40, 19., -25]

print("표준편차 of set1 :%9.3f" % (math.pow(variance(myset1), 1/2)))
print("표준편차 of set2 :%9.3f" % (math.pow(variance(myset2), 1/2)))

print("Variance of set1 :%9.3f" % variance(myset1))
print("Variance of set2 :%9.3f" % variance(myset2))

fig = plt.figure()

sub1 = fig.add_subplot(2,1,1)
sub1.axis((0,len(myset1)-1,-50,20))
sub1.plot(myset1, 'g')
sub1.plot(myset1, 'ro')
sub1.plot([mean(myset1)]*len(myset1), 'b--')
sub1.plot([mean(myset1)+(math.pow(variance(myset1), 1/2))]*len(myset1), 'y--')
sub1.plot([mean(myset1)-(math.pow(variance(myset1), 1/2))]*len(myset1), 'y--')

sub2 = fig.add_subplot(2,1,2)
sub2.axis((0,len(myset2)-1,-50,20))
sub2.plot(myset2, 'g')
sub2.plot(myset2, 'ro')
sub2.plot([mean(myset2)]*len(myset2), 'b--')
sub2.plot([mean(myset2)+(math.pow(variance(myset2), 1/2))]*len(myset2), 'y--')
sub2.plot([mean(myset2)-(math.pow(variance(myset2), 1/2))]*len(myset2), 'y--')


plt.show()
```

축의 **axis(x1, x2, y1, y2)**로 표현되는 범위를 조정할수 있다.  
**x1, x2, y1, y2 = getAxis()** 하면 4개의 값이 있는 리스트가 반환되어서 값을 확인할 수 있다.

## 히스토그램으로 정규분포 나타내기

    히스토그램은 데이터의 갯수를 나타내는 그래프이다.
    히스토그램의 y축값을 총갯수로 나누면 확률 분포가 된다.
    plt.hist(data, bins=100)
    bins는 x축에 표시되는 구간의 갯수를 의미한다.

    정규분포 random.normal(0,1,1000)는
    평균이 0, 표준편차 1인 정규분포를  따르는 변수 1000개 생성.


```python
import numpy as np
data = np.random.normal(0, 1, 10000)
import matplotlib.pyplot as plt

plt.hist(data, bins = 50, label='bin=50')
plt.legend(loc='upper left')
plt.show() 

plt.hist(data, bins = 10, label='bin=10')
plt.legend(loc='upper left')
plt.show()
```

## 히스토그램으로 균등 분포(Uniform Distribution) 나타내기

    균등 분포는 정규분포와 다르게 데이터가 균일하게 분포한다.
    np.random.rand(50000)
    np.random.rand(1000)
    이런식으로 데이터를 몇개 생성할지를 입력변수로 쓸수 있다.

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.rand(50000)
plt.hist(data, bins=100, color = 'k')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
new1 = fig.add_subplot(2,2,1)

data = np.random.rand(10)
new1.hist(data, bins=10)

new2 = fig.add_subplot(2,2,2)
data = np.random.rand(100)
new2.hist(data, bins=100)

new2 = fig.add_subplot(2,2,3)
data = np.random.rand(1000)
new2.hist(data, bins=100)

new3 = fig.add_subplot(2,2,4)
data = np.random.rand(100000)
new3.hist(data, bins=100, color = 'r')
plt.show()
```
